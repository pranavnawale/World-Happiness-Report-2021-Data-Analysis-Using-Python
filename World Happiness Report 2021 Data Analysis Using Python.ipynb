# World Happiness Report 2021 Data Analysis Using Python


## Dataset Link

```python
# install the dependencies
import pandas as pd
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
```

```python
happy_report=pd.read_csv("C:\\Users\\Pranav\\Desktop\\DATA SCIENCE DATA\\CVC file\\happines_report\\world-happiness-report-2021.csv")
happy_report.head()
```

```python
data_columns=['Country_name','Regional_indicator','Logged_GDP-per-capita','Happiness_score','Social-support','Healthy_life_expectancy','Freedom_to_make_life_choices','Generosity','Perceptions_of_corruption']
```

```python
happy_report=happy_report[data_columns].copy()
```

```python
happy_report.head()
```

```python
# shape of dataset
happy_report.shape
```

```python
# information about dataset
happy_report.info()
```

```python
#Describe of dataset
happy_report.describe()
```

```python
# find ot null value in dataset
happy_report.isnull().sum()
```

```python
# Scatterplot
plt.rcParams['figure.figsize']=(15,7)
plt.title('Plot between Happiness Score and GDP')
sns.scatterplot(x='Happiness_score',y='Logged_GDP-per-capita',hue='Regional_indicator',s=200,data=happy_report);
plt.legend(loc="upper left",fontsize='10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP_per_capita')
plt.show()
```

```python
#GDP as per region 
GDP_region=happy_report.groupby('Regional_indicator')['Logged_GDP-per-capita'].sum()
```

```python
GDP_region
```

```python
# pie plot
GDP_region.plot.pie(autopct='%1.1f%%')
plt.title(" GDP by Region")
plt.ylabel('')
plt.show()
```

```python
# total countries
total_country=happy_report.groupby('Regional_indicator')[['Country_name']].count()
print(total_country)
```

```python
# Correlation heatmap
corr=happy_report.corr(method="pearson")
f,ax=plt.subplots(figsize=(10,5))
sns.heatmap(corr,mask=np.zeros_like(corr,dtype=np.bool),cmap="Blues",square=True,ax=ax)
plt.show()          
```

```python
# Corruption in region
corruption=happy_report.groupby('Regional_indicator')[['Perceptions_of_corruption']].mean()
corruption
```

```python
# Barplot
plt.rcParams['figure.figsize']=(12,8)
plt.title('Percepation of Corruption in Various Region')
plt.xlabel('Rrgions',fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha="right")
plt.bar(corruption.index,corruption.Perceptions_of_corruption)
plt.show()   
```

```python
top_10=happy_report.head(10)
bottom_10=happy_report.tail(10)
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))
plt.tight_layout(pad=2)
xlabels=top_10.Country_name
axes[0].set_title('Top_10 happiest countries Life Expetancy')
axes[0].set_xticklabels(xlabels,rotation=45,ha="right")
sns.barplot(x="Country_name",y="Healthy_life_expectancy",data=top_10,ax=axes[0])
axes[0].set_xlabel("Country Name")  
axes[0].set_ylabel("Life Expectancy")

xlabels=bottom_10.Country_name
axes[0].set_title('Bottom_10 happiest countries Life Expetancy')
axes[0].set_xticklabels(xlabels,rotation=45,ha="right")
sns.barplot(x="Country_name",y="Healthy_life_expectancy",data=bottom_10,ax=axes[1])
axes[0].set_xlabel("Country Name")  
axes[0].set_ylabel("Life Expectancy")

plt.grid(True)
plt.show()
```

```python
# Scatterplot
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x="Freedom_to_make_life_choices",y="Happiness_score",data=happy_report,hue="Regional_indicator")
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```

```python
country=happy_report.sort_values(by='Perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize']=(12,8)
plt.title('Countries with most perception of corruption')
plt.xlabel('Country',fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha="right")
plt.bar(country.Country_name,country.Perceptions_of_corruption)
plt.show()   
```

```python
# Scatterplot
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x="Happiness_score",y="Perceptions_of_corruption",data=happy_report,hue="Regional_indicator")
plt.xlabel('Happiness_score')
plt.ylabel('Perceptions_of_corruption')
plt.grid(True)
plt.show()
```
